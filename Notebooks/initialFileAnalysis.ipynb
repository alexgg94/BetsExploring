```python
import json
import os.path
import sys
import csv
import numpy as np
import pandas as pd
```

```python
pathori = "../Data/Raw/"
pathtemp = "../Data/Interim/"
pathdest = "../Data/Processed/"
```

```python
season_list = []
for folder in os.listdir(pathori):
    country = folder
#    print(pathori+country)
    for file in os.listdir(pathori+folder):
        competition = file.split("_")[0]
        season_short = (file.split("_")[1]).split(".")[0]
        season = "20" + season_short[0:2] + "-" + "20" + season_short[2:4]
        season_info = {"country": country, "competition": competition, "season":season}
        season_list.append(season_info)
        
df = pd.DataFrame(season_list)
print df.head(20)
```
