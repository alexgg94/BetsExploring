## 5_fieldsAnalysis - recent seasons dataset

In this notebook we will create a subdataset with the most recent seasons (2017/18 - 2018/19)

```python
import os
import os.path
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from colorama import Fore
```

```python
pathori = "../Data/Raw/Main/"
pathtemp = "../Data/Interim/"
pathdest = "../Data/Processed/"
filetemp = "main_competitions.csv"
filedest = "main_competitions.csv"
filerecent = "main_competitions_recent.csv"
```

```python
# Open dataset
url = pathdest + filedest
parse_dates=["Date"]
ds = pd.read_csv(url, parse_dates=parse_dates, index_col=False)
```

```python
print("Shape: ", ds.shape)
print("Columns:", ds.columns)
print("Sample data:")
display(ds.sample(5))
```

```python
dsrecent = ds.loc[ds['Season']>='2017-2018']
print("Shape:", dsrecent.shape)
```

```python
# Detect columns with nulls
dsrecent.isnull().sum()
```

```python
# Drop quota columns with nulls in all records
# BSH, BSD, BSA, GBH, GBD, GBA, SBH, SBD, SBA, SJH, SJD, SJA
dsrecent = dsrecent.drop(labels=['BSH', 'BSD', 'BSA', 'GBH', 'GBD', 'GBA', 'SBH', 'SBD', 'SBA', 'SJH', 'SJD', 'SJA'],axis=1)
```

```python
print (dsrecent.shape)
print (dsrecent.columns)
display(dsrecent.sample(5))
```

```python
url = pathdest + filerecent
if os.path.exists(url):
    os.remove(url)
    
dsrecent.to_csv(url, index=None)
```

```python
aggregation = {"Season": ["min", "max", "count"]}
dsrecent.groupby(["Country","Competition"]).agg(aggregation)
```

```python
# Number of matches with quota. aggregate by Contry and Competition
dsrecent.pivot_table(index=['Country','Competition'],columns=[],values=['B365H','BWH','IWH','LBH','VCH','WHH'],aggfunc='count',fill_value=0)
```
