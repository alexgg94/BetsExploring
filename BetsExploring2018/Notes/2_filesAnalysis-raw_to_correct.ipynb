We also will analyse fields in every file to select final fields to be included in dataset

```python
import json
import os.path
import sys
import csv
import numpy as np
import pandas as pd
```

```python
pathori = "../Data/Raw/Main/"
pathtemp = "../Data/Interim/"
pathdest = "../Data/Processed/"
```

```python
season_list = []
cols_list = []

#print(pathori)
for folder in os.listdir(pathori):
    country = folder
    current_compet = ""
#    print ('  ' + country)
    for subfolder in os.listdir(pathori+folder):
        competition = subfolder
#        print ('    '+competition)
        
        for file in os.listdir(pathori+'/'+folder+'/'+subfolder):
            season_short = file.split(".")[0]
            season = "20" + season_short[0:2] + "-" + "20" + season_short[2:4]
#            print ('      '+season)
            season_info = {"country": country, "competition": competition, "season":season}
        
            # Store columns information
            file_path = pathori+folder+"/"+subfolder+"/"+file
#            print (file_path)
            try:
                ds = pd.read_csv(file_path)
                season_list.append(season_info)   
            
                if competition != current_compet:
                    current_compet = competition
                
                    cols = list(ds)
                    for col in cols:
                        cols_list.append({"country":country, "competition":competition, "column": col})
                 
            except:
                print ("Can't read file: " + file_path)
            
```

```python
dsfiles = pd.DataFrame(season_list)
display (dsfiles.sample(10))
```

```python
dsfiles.describe()
```

```python
dscols = pd.DataFrame(cols_list).sort_values("column")
display (dscols.head(5))
```

```python
aggregation = {"season": ["min", "max", "count"]}
dsfiles.groupby(["country","competition"]).agg(aggregation)
```
