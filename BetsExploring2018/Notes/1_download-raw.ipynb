# Data Mining project
## 1_download - Raw
In this phase of the project we will download original files and we will move in our convenient structure of folders and files

```python
import os
import shutil
import urllib3
import pandas as pd
from colorama import Fore
```

```python
# Global variables

index_url="../Notes/file_index.csv"
main_base_url = "http://www.football-data.co.uk/mmz4281/"
extra_base_url = "http://www.football-data.co.uk/new/"
    
main_dest_base_url = "../Data/Raw/Main/"
extra_dest_base_url = "../Data/Raw/Extra/"

seasons = [ "0304", "0405", "0506", "0607", 
            "0708", "0809", "0910", "1011", 
            "1112", "1213", "1314", "1415", 
            "1516", "1617", "1718", "1819" ]
```

```python
# Destination folders creations

def initFolders():
    if os.path.exists(main_dest_base_url):
        shutil.rmtree(main_dest_base_url)
    os.makedirs(main_dest_base_url)
    
    if os.path.exists(extra_dest_base_url):
        shutil.rmtree(extra_dest_base_url)
    os.makedirs(extra_dest_base_url)
```

```python
# Download one original file with destination name

def downloadFile(file_ori, file_dest):
    http = urllib3.PoolManager()
    r = http.request('GET', file_ori, preload_content=False)
    response = r.read()
    fout = open(file_dest,"wb")
    fout.write(response)
    fout.close()
    
```

```python
# Download files in destination folders

def downloadFiles():
    
    index = []
    columns = ['ori_folder','ori_file','dest_folder','dest_file']
    result = pd.DataFrame(index=index, columns=columns)
    
    ds = pd.read_csv(index_url) # Index of competitions
    for index, row in ds.iterrows():
        if row['type'] == 'M':  # Main competitions
            for season in seasons:
                file_ori_url = main_base_url + season + '/' + row['file'] + '.csv'
                file_dest_url = main_dest_base_url + row['country'] + '/' + row['competition'] + "/" + season + ".csv"
                
                if not os.path.exists(main_dest_base_url + row['country']):
                    os.makedirs(main_dest_base_url + row['country'])
                if not os.path.exists(main_dest_base_url + row['country'] +  '/' + row['competition']):
                    os.makedirs(main_dest_base_url + row['country'] +  '/' + row['competition'])
                    
                print(Fore.BLUE + "Reading: " + row['country'], row['file'], row['competition'], season)
                print(Fore.GREEN + file_ori_url) 
                print(Fore.GREEN + "-> " + file_dest_url)

                try:
                    downloadFile(file_ori_url, file_dest_url)
                    line = pd.DataFrame([[season ,row['file'] + '.csv',row['country'] + '/' + row['competition'],season + ".csv"]],  columns=columns)
                    result = result.append(line)

                except:
                    print(Fore.RED + "Could not download " + file_ori_url) 
                    
        else: # Extre competitions
            file_ori_url = extra_base_url + row['file'] + '.csv'
            file_dest_url = extra_dest_base_url + row['country'] + '/' + row['competition'] + ".csv"
            print(Fore.BLUE + "Reading: " + row['country'], row['file'], row['competition'])
            print(Fore.GREEN + file_ori_url) 
            print(Fore.GREEN + "-> " + file_dest_url)

            if not os.path.exists(extra_dest_base_url + row['country']):
                os.makedirs(extra_dest_base_url + row['country'])

            try:
                downloadFile(file_ori_url, file_dest_url)
                line = pd.DataFrame([['',row['file'] + '.csv',row['country'], row['competition'] + ".csv"]],  columns=columns)
                result = result.append(line)
                
            except:
                print(Fore.RED + "Could not download " + file_ori_url) 
                
    print(result)
    print(Fore.YELLOW + "Download finished")
```

```python
# Main script
initFolders()
downloadFiles()
```
